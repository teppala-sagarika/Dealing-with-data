```python
import numpy as np
import time
```

```python
arr=np.array([1,2,3,4,5])
print(arr,type(arr))
```

```python
#comparing execution performance
size=100_000_000

#pylist
py_list=list(range(size)) #[0,1,2,...size-1]
start=time.time()
sq_list=[x**2 for x in py_list]
end=time.time()
print(f"py list time = {end-start}")

#numpy 
np_arr=np.array(py_list)
start=time.time()
sq_arr=np_arr**2 #vectorization
end=time.time()
print(f"np array time = {end-start}")
```

```python
#memory 
import sys
print(f"py list size = {sys.getsizeof(py_list)*len(py_list)} bytes")
print(f"numpy arr size = {np_arr.nbytes} bytes")
```
